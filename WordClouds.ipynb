```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator


```

```python
with open('Ogaan.txt','r') as file:
    content = file.read()
```

```python
content = " ".join(content.split())
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS, width=1000, height=800, background_color='white',font_path='./arial.ttf')
wordcloud.generate(content)

plt.title("Ogaan")
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()

```

The Ogaan Word cloud is the least dense of the three word clouds. Probably because the content on the website is
lesser as opposed to the other two websites. The website probably has handpicked labels for people who are
looking to wear designers but are also have a preference for a category more than a designer.
